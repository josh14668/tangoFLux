```python
class AudioDataset(Dataset):
    def __init__(self, manifest_path:str,
                 sample_rate: int = 44100,
                 duration: float = 10.0):
      with open(manifest_path, "r") as f:
        self.entries = json.load(f)
      self.sample_rate = sample_rate
      self.duration = duration
      self.num_samples = int(self.sample_rate * self.duration)

    def __len__(self):
        return len(self.entries)

    def __getitem__(self, idx):
        path = self.entries[idx]['location']

        wav,sr = librosa.load(path,
                              sr = self.sample_rate,
                              mono=True,
                              dtype='float32')

        if wav.shape[0] > self.num_samples:
          pad = self.num_samples - wav.shape[0]
          wav = F.pad(torch.from_numpy(wav),
            (0, pad),
            mode='constant',
            value=0.0)
        else:
            wav = torch.from_numpy(wav[:self.num_samples])


        return wav
```

```python
SR       = 44100
DURATION = 10            # seconds (same as TangoFlux)
SAMPLES  = SR * DURATION

device = torch.device("cuda")


sc = Scattering1D(J=10, shape=SAMPLES, Q=8).to(device)   # typical J,Q
wav = torch.randn(1, SAMPLES).to(device)                 # dummy audio
with torch.no_grad():
    S = sc(wav)               # (B, C_scatter, T_scatter)

C_scatter, T_scatter = S.shape[-2:]
print("Scattering dims:", C_scatter, T_scatter)
```

```python
class Scatter2Latent(nn.Module):
    def __init__(self, C_scatter, T_scatter, latent_ch=64, kernel=1, T_lat=645):
        super().__init__()

        # 1×1 conv to map each scatter‐vector → latent‐vector
        # self.proj = nn.Conv1d(
        #     in_channels=C_scatter,
        #     out_channels=latent_ch,
        #     kernel_size=kernel,
        #     padding=kernel//2
        # )

        self.proj = nn.Sequential(
            nn.Conv1d(
                in_channels=C_scatter,
                out_channels=latent_ch,
                kernel_size=kernel,
                padding=kernel//2
            ),

            nn.ReLU(),

            nn.Conv1d(
                in_channels=128,
                out_channels=64,
                kernel_size=3,
                padding=kernel//2
            ),

            nn.ReLU(),

            nn.Conv1d(
                in_channels=64,
                out_channels=1,
                kernel_size=3,
                padding=kernel//2
            )
        )

        self.need_resample = (T_scatter != T_lat)
        if self.need_resample:
            # interpolate time from T_scatter → T_lat
            self.interp = lambda x: F.interpolate(
                x, size=T_lat, mode="linear", align_corners=False
            )

    def forward(self, S):
        # S: (B, T_scatter, C_scatter)
        x = S.permute(0, 2, 1)   # → (B, C_scatter, T_scatter)
        x = self.proj(x)         # → (B, latent_ch, T_scatter)
        if self.need_resample:
            x = self.interp(x)   # → (B, latent_ch, T_lat)
        return x                 # → (B, latent_ch, T_lat)
```

```python
vae = AutoencoderOobleck.from_pretrained("stabilityai/stable-audio-open-1.0",
                                         subfolder="vae")
decoder = vae.decoder.eval().requires_grad_(False).to(device)
```

```python
dataset = AudioDataset("/content/TangoFlux/data/train.json",
                       sample_rate=44100,
                       duration=10.0)
loader = DataLoader(dataset, batch_size=1, shuffle=True)

proj = Scatter2Latent(
    C_scatter=431,   # in_channels
    T_scatter=374,   # time frames coming out of Scattering1D
    latent_ch=64,    # the VAE’s channel count
    kernel=1,        # per‑time‐step linear mapping
    T_lat=645        # target latent time‐length
).cuda()

optim    = torch.optim.Adam(proj.parameters(), lr=1e-1)
loss_fn  = nn.MSELoss()
epochs   = 50


spec_transform = torchaudio.transforms.Spectrogram(
    n_fft=1024,
    hop_length=256,
    power=1.0  # returns magnitude
).to(device)

# 1) Outer loop with epoch bar
for epoch in range(epochs):
    epoch_bar = tqdm(loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False)
    for wav in epoch_bar:
        wav   = wav.cuda()
        mono  = wav.mean(1) if wav.dim()==3 else wav    # (B,N)

        S     = sc(mono)                                # (B, T_sc, C_sc)
        z     = proj(S)                                 # (B,64,645)

        # decode through the full VAE so grads flow into proj only
        #with torch.no_grad():
        recon = vae.decode(z).sample                    # (B,2,N_dec)

        # 5) slice to match your target length
        Tgt = mono.shape[-1]
        if recon.size(-1) != Tgt:
            recon = recon[..., :Tgt]

        # 6) collapse to mono / pick one channel
        pred   = recon[:, 0, :]
        target = mono




        # --- waveform loss ---
        wav_loss = F.mse_loss(pred, target)

        # --- spectral loss ---
        # Spectrogram expects (B, channel, time), so unsqueeze(1)
        S_pred   = spec_transform(pred.unsqueeze(1))    # (B, freq, time)
        S_target = spec_transform(target.unsqueeze(1))
        spec_loss = F.l1_loss(S_pred, S_target)

        # --- combine ---
        loss = 0.5 * wav_loss + 0.5 * spec_loss


        optim.zero_grad()
        loss.backward()
        optim.step()

        # update the inner bar’s postfix with loss
        epoch_bar.set_postfix(loss=loss.item())
    # you can print a summary per‐epoch
    print(f">>> End epoch {epoch+1}, last loss: {loss.item():.6f}")
```

```python
import torchaudio
from kymatio.torch import Scattering1D

device = torch.device("cuda")
# 1) Load your WAV (mono) and pick a fixed length T
wav, sr = torchaudio.load("/content/TangoFlux/data/test.wav", normalize=True)  # (1 or 2, N)
mono = wav.mean(0, keepdim=True)                           # (1, N)
T = mono.shape[-1]

# 2) Build your scatterer to match T
J, Q = 8, 8          # or whatever you chose originally
scattering = Scattering1D(J=J, shape=T, Q=Q).to(device)

# 3) Compute target scattering once
with torch.no_grad():
    Sx = scattering(mono.to(device)).detach()  # (1, C_scatter, T_scatter)

# 4) Initialize y and run the same optimization loop
y = torch.randn_like(mono, requires_grad=True, device=device)  # (1, T)
optimizer = torch.optim.Adam([y], lr=1e-1)

history = []
for step in range(200):
    Sy = scattering(y)
    loss = F.mse_loss(Sy, Sx)
    optimizer.zero_grad(); loss.backward(); optimizer.step()
    history.append(loss.item())

# 5) y now holds your reconstructed waveform—write it back out
recon = y.detach().cpu()
torchaudio.save("reconstruction.wav", recon, sr)
```

```python
# 1) Load your original WAV
wav, sr = torchaudio.load("/content/TangoFlux/data/test.wav")  # (2,N), 44100

# 2) Run it through the same pipeline:
mono = wav.mean(0, keepdim=True).cuda()  # (1,N)
proj.eval()

# Resample the audio to match the expected length of the scattering transform
mono = torchaudio.functional.resample(mono, sr, 44100) # Resample to 44100 Hz
mono = F.pad(mono, (0, 441000 - mono.shape[-1]), mode='constant', value=0) # Pad to 441000 samples

S    = sc(mono)                          # (1,374,431)
z    = proj(S)                           # (1,64,645)
recon = vae.decode(z).sample[0].cpu()    # (2,N_recon)

# 3) Slice/pad to original length
recon = recon[..., : wav.shape[1]]

# 4) Save & play back
torchaudio.save("recon.wav", recon, sr)
Audio("recon.wav")
```
